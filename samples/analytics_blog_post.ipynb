# Altair + Jupyter Notebooks vs Tensorboard?

I am a heavy user of [Tensorboard](https://www.tensorflow.org/tensorboard).
However, there were many instances where I had to do custom visualizations.
I used to visualize on [notebooks](https://github.com/vpj/lab/blob/2.0/analytics/sample_analytics.ipynb)
with `matplotlib`.

Recently we came across [Altair](https://altair-viz.github.io/).
It's a Python API to create [Vega](https://vega.github.io/vega/) schema for visualizations,
 and Vega renders interactive charts with [d3](https://d3js.org/).
 
We are thinking of completely switching over to Altair + Notebooks.
The two main questions are:
1. Which is better?
2. Which is easier?

```python
from lab.analytics import Analyzer

# a = Analyzer('mnist_loop', '9070f0ba756511ea8c99acde48001122')
a = Analyzer('mnist_loop', '1d3f855874d811eabb9359457a24edc8')

Histograms, Scalars = a.get_indicators('Histogram', 'Scalar')

# print(dir(Scalars))
# print(dir(Histograms))
```

We can use [EventAccumulator](https://github.com/tensorflow/tensorboard/blob/master/tensorboard/backend/event_processing/event_accumulator.py) to extract data from Tensorboard.

```python
a.tb.load()
```

I mostly use Tensorboard's Scalars and Distributions tabs.

Scalars can be easily replicated on a notebook.

```python
a.tb.render_scalar_simple(Scalars.train_loss, points=None).interactive()
```

Here's a smoothed out version, with variance bands, which I tend to prefer.

```python
a.tb.render_scalar_simple(Scalars.train_loss).interactive()
```

The one I use is smoothed out with a view finder.
It is so much easier to zoom and go into details of a chart with it.

```python
a.tb.render_scalar(Scalars.train_loss)
```

Histograms can also be visualized on a notebook just like with Tensorboard.
And you can zoom into them, which you can not do on Tensorboard.
You can also have a view finder.

```python
a.tb.render_histogram(Histograms.model_conv1_weight_grad)
```

The best thing about visualizing on notebooks is that you can customize.
For instance, here's two histograms laid on top of each other. 
Not a very useful visualization here, but you get the point.

```python
a.tb.render_histogram_multiple([Histograms.model_fc1_weight_grad, Histograms.model_conv1_weight_grad])
```

**So the visualizations on notebooks are clearly better.**
What about the ease of use.
Tensorboard is definety easier to use.
But visualizing with a single line of code isn't so bad.

We haven't used this extensively yet.

 the frequently needed visualizations.
